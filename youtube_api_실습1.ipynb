```python
from youtube_transcript_api import YouTubeTranscriptApi

# Youtube 자막 추출 함수
def get_transcript(video_id, languages=['ko']):
  transcript = YouTubeTranscriptApi.get_transcript(video_id, languages=['ko'])
  full_transcript = ' '.join([item['text'] for item in transcript])
  return full_transcript
```

```python
# Video ID 추출
video_id = video_list['items'][3]['id']['videoId']

# Youtube 영상의 자막 받아오기
full_transcript = get_transcript(video_id)
full_transcript
```

```python
import requests
import json

# API 정보
WIKI_API_KEY = "zautomztovfzjkdzcporbgkabffzxx"
url = 'http://www.wikifier.org/annotate-article'

params = {
    'text': full_transcript,
    'lang': 'ko',
    'userKey': WIKI_API_KEY,
    'pageRankSqThreshold': 0.8,
    'applyPageRankSqThreshold': 'true',
    'nTopDfValuesToIgnore': 100,
    'nWordsToIgnoreFromList': 100
}

response = requests.post(url, data=params)
wikis = response.json()

wikis
```

```python
keywords = set()
for d in wikis['ranges']:
  for candidate in d['candidates']:
    if candidate['cosine'] > 0.05:
      keywords.add(candidate['title'])
keywords
```

```python
# 1. 한글 폰트 설치
!apt-get install -y fonts-nanum
!fc-cache -fv
!rm ~/.cache/matplotlib -rf

# 2. 폰트 설정
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm
import matplotlib as mpl
import networkx as nx
from itertools import combinations

# 나눔고딕 경로 설정
font_path = "/usr/share/fonts/truetype/nanum/NanumBarunGothic.ttf"
font_prop = fm.FontProperties(fname=font_path)

# matplotlib 전역 설정 (가능한 범위에서)
mpl.rc('font', family=font_prop.get_name())
plt.rcParams['axes.unicode_minus'] = False

# 3. 영상 키워드 예시 (공통 키워드 포함)
video_keywords = {
    "영상 A": {'광고',
 '달 (점성술)',
 '말 (동물)',
 '블로그',
 '비디오 게임 개발자',
 '생명',
 '소프트웨어',
 '소프트웨어 개발자',
 '시간',
 '여러분 (노래)',
 '월드 와이드 웹',
 '웹 게임',
 '웹 애플리케이션',
 '인공지능',
 '인과관계',
 '정보',
 '주식 투자',
 '진입 장벽',
 '프로그래머'},
    "영상 B": {'말 (동물)', '소프트웨어 개발자', '시간', '인공지능', '컴파일러', '컴퓨터', '피아제의 인지 발달론'},
    "영상 C": {'AI (가수)',
 '기술적 특이점',
 '로봇',
 '로봇 (기업)',
 '로봇공학',
 '소프트웨어',
 '소프트웨어 개발',
 '소프트웨어 개발 프로세스',
 '소프트웨어 개발자',
 '소프트웨어 공학',
 '소프트웨어 공학자',
 '스마트폰',
 '시간',
 '응용 소프트웨어',
 '인공지능',
 '조류 인플루엔자',
 '컴퓨터 산업',
 '투자',
 '프로그래머'},
}

# 4. 그래프 생성
G = nx.Graph()
G.add_nodes_from(video_keywords.keys())

for vid1, vid2 in combinations(video_keywords.keys(), 2):
    shared = video_keywords[vid1] & video_keywords[vid2]
    if shared:
        G.add_edge(vid1, vid2, weight=len(shared), label=", ".join(shared))

# 5. 시각화
pos = nx.spring_layout(G, seed=42)

plt.figure(figsize=(10, 6))
nx.draw_networkx_nodes(G, pos, node_size=1000, node_color="lightyellow")

# 🔥 한글 폰트 적용 중요 포인트
nx.draw_networkx_labels(G, pos, font_size=12, font_family=font_prop.get_name())
nx.draw_networkx_edges(G, pos, width=2)

edge_labels = nx.get_edge_attributes(G, "label")
nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels, font_size=10, font_family=font_prop.get_name())

plt.title("유튜브 영상 간 공통 키워드 네트워크", fontsize=14, fontproperties=font_prop)
plt.axis("off")
plt.show()
```
